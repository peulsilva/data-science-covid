```python
import pandas as pd 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("20211216_Casos-e-obitos-ESP.csv", sep =";")
df
```

```python
df = df.dropna()
contaminados = df[df["Diagnostico Covid19"]== "CONFIRMADO"]
```

```python
INDEFINIDO = 0
IGNORADO = 1
MASCULINO = 2
FEMININO = 3
contaminados.loc[:,'Genero']=contaminados.loc[:,'Genero'].replace('IGNORADO',IGNORADO)
contaminados.loc[:,'Genero']=contaminados.loc[:,'Genero'].replace('INDEFINIDO',INDEFINIDO)
contaminados.loc[:,'Genero']=contaminados.loc[:,'Genero'].replace('MASCULINO',MASCULINO)
contaminados.loc[:,'Genero']=contaminados.loc[:,'Genero'].replace('FEMININO',FEMININO)
contaminados['Genero'].value_counts()
```

```python
x= contaminados[["Genero", "Idade" ]]
y = contaminados["Obito"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(random_state=1)
rfc.fit(x_train,y_train)
```

```python
pred=rfc.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, pred))
print('\n')
print("=== Classification Report ===")
print(classification_report(y_test, pred))
print('\n')
```
